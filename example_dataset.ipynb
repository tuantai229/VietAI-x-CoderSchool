```python
from torchvision.datasets import CIFAR10
from torch.utils.data import Dataset, DataLoader
import os
import pickle
import numpy as np
import cv2

# dataset = CIFAR10(root="my_dataset", train=True, download=True)
# image, label = dataset[1000]
# image.show()
# print(dataset.classes[label])

class CIFAR10Dataset(Dataset):
    def __init__(self, data_path, train):
        if train:
            data_files = [os.path.join(data_path, "data_batch_{}".format(i)) for i in range(1, 6)]
        else:
            data_files = [os.path.join(data_path, "test_batch")]
        # print(data_files)

        self.images = []
        self.labels = []
        for data_file in data_files:
            with open(data_file, "rb") as fo:
                # print(fo)
                data = pickle.load(fo, encoding="latin1")
                # print(data)
                # print(type(data))
                # print(data.keys())
                # print(data["batch_label"]) #ko cần
                # print(data["labels"]) #cần
                # print(data["data"]) #cần
                # print(data["filenames"]) #ko cần
                self.images.extend(data["data"])
                self.labels.extend(data["labels"])
    
    def __len__(self):
        return len(self.labels)

    def __getitem__(self, item):
        image = self.images[item]
        image = np.reshape(image, (3, 32, 32))
        label = self.labels[item]
        return image, label


if __name__ == '__main__':
    path = "my_dataset/cifar-10-batches-py"
    dataset = CIFAR10Dataset(data_path=path, train=True)
    # print(len(dataset))
    # image, label = dataset[1000]
    # image.show()
    # print(type(image))

    # image = np.reshape(image, (3, 32, 32))
    # image = np.transpose(image, (1, 2, 0))
    # print(label)
    # cv2.imshow("test", cv2.resize(image, (320, 320)))
    # cv2.waitKey(0)

    dataloader = DataLoader(
        dataset=dataset,
        batch_size=16,
        shuffle=True,
        num_workers=0,
        drop_last=True
    )
    for images, labels in dataloader:
        print(images.shape, labels.shape)
```
